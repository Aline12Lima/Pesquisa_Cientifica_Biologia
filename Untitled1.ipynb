```python
# Importa a função de processamento de imagem do seu arquivo
from processador import extrair_caracteristicas
from sklearn.svm import SVC
import numpy as np

# 1. Usa a função para pegar os dados reais da sua imagem
caminho_da_imagem = 'exemplo.jpg'
dados_reais = extrair_caracteristicas(caminho_da_imagem)

if dados_reais is None or not dados_reais:
    print("Nenhum dado real foi extraído da imagem. O modelo não pode fazer previsões.")
else:
    # Transforma os dados em um array NumPy para o ML
    X_teste = np.array(dados_reais)

    # 2. Treinamento do modelo com os dados de exemplo
    X_treino = np.array([[500], [800], [300], [450], [650], [250], [750]])
    y_treino = np.array([0, 1, 0, 1, 0, 0, 1])

    modelo = SVC()
    modelo.fit(X_treino, y_treino)

    # 3. Faz a previsão com os dados reais
    previsoes = modelo.predict(X_teste)

    print("Características reais extraídas:", X_teste)
    print("Previsões do modelo:", previsoes)
```
